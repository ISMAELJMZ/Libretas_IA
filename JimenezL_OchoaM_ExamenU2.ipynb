# parte 2, modelo de clasificacion 

```python
# importamos la librerias para el arbol de decision 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import pandas as pd 
```

```python
# importamos los datos para el arbol de decision, este caso es el data2
Data2 = pd.read_csv('data2.csv')
Data2
```

```python
X = Data2.iloc[:,:-1]
y = Data2[["Intensity"]]
```

```python
#entrenamos los datos 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6,random_state=45)
```

```python
# creamos el modelo basico 
modelo_dt = RandomForestClassifier()
modelo_dt.fit(X_train, y_train)
modelo_dt.score(X_test, y_test)
```

# breve descripcion del algoritmo y suposiciones 
se utiliza el arbol de desicion cuando queremos clasificar nuevo datos y el resultado que se mustra es la desicion que se tomo sobre cual nodo es el que se debe elegir

```python
# validacion cruzada infromando la media y desviacion estandar 
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV

grid = {
    'n_estimators' : [200,300,400,500],
    'max_features' : ['sqrt','log2'],
    'max_depth'    : [4,5,6,7,8],
    'criterion'    : ['gini','entropy'],
    'random_state': [18] 
    
}
rf_cv = GridSearchCV(estimator = RandomForestClassifier(), param_grid = grid, cv=5)
rf_cv.fit(X_train, y_train)


scores = cross_val_score(modelo_dt, X, y, cv=5)

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))


```

# explicacion de porque es importante la validacion cruzada 

la validacion cruzada nos permite evaluar el rendimiento de una forma un tanto mas precisa y confiable que dividir los datos en un conjunto de entrenamiento y uno de prueba, es de alguna manera mas confiable al predecir

```python
# verificamos cuales son los mejores parmetros para ajustarlos 
rf_cv.best_params_
```

```python
#ajustamos los nuevos parametros 
rf2 = RandomForestClassifier(n_estimators = 200,
                             max_depth    = 4,
                             max_features = 'sqrt',
                             random_state = 18,
                             criterion    = 'gini').fit(X_train,np.ravel(y_train))
```

```python
# puntuacion de la validacion cruzada 
y_pred = rf2.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

# aplique el significado de las metricas de precision, recupreacion y puntaje F1

son meidas comunes que se utilizan ppara evaluar el rendimiento de un modelo de clasificacion, estas metricas se calculan a partir de la matriz de confusión, que es una tabla que resume el numero de verdaderos positivos, verdaderos negativos, falsos positivos y falsos negativos que produce el modelo.



```python
# modelo ajustado con hiperparametros 
modelo_dt = RandomForestClassifier(criterion= "gini",max_depth=4,max_features="sqrt",n_estimators= 200,random_state = 18)
modelo_dt.fit(X_train, y_train)
modelo_dt.score(X_test, y_test)
```
